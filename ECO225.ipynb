```python
import pandas as pd
```

```python
biden = pd.read_csv('Documents/hashtag_joebiden.csv')
```

```python
#Trump file is larger, so read in chunks
chunks = pd.read_csv('Documents/hashtag_donaldtrump.csv', chunksize = 100)
trump = pd.concat(chunks)

df = pd.concat([biden, trump], axis=0)
```

```python
#Converting into Dataframe
pd.DataFrame(df)
```

```python
#Dropping the missing values from all columns
df.dropna()
```
